```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
def binary_tree(num,node):
    if num<node.data:
        if not node.left:
            node.left=Node(num)
        else:
            binary_tree(num,node.left)
    else:
        if not node.right:
            node.right=Node(num)
        else:
            binary_tree(num,node.right)

def show_in_order(node):
    if not node:
        return
    show_in_order(node.left)
    print(node.data)
    show_in_order(node.right)
    
    
    
head=Node(6)
list_of_nums=[10, 2, 4, 5, 12]
for i in list_of_nums:
    binary_tree(i,head)

    
def delete_node(data,head):
    node=head
    #head 가 유일한 노드인경우
    if node.left==node.right and node.right== None:
        if node.data==data:
            node=None
        else:
            print("no such thing(1)")
        return
    #찾기
    left=True
    while node:
        if node.data==data:
            break
        elif data<node.data:
            node=node.left
            left=True
        else:
            node=node.right
            left=False
    #교체
    top_node=node
    
    while not (node.left==node.right and node.right==None):
        print("----")
        show_in_order(head)
        print("----")
        top_node=node
        prev_node=node
        if node.left!=None:
            prev_node=node
            node=node.left
            left=False
            while node.right!=None:
                prev_node=node
                node=node.right
            top_node.data=node.data
        else:
            prev_node=node
            node=node.right
            left=True
            while node.left!=None:
                prev_node=node
                node=node.left
            top_node.data=node.data
            
    print("----")
    show_in_order(head)
    print("----")
    print(previous_node.data, "<<<pre-check")
    if left:
        previous_node.left=None
    else:
        previous_node.right=None
        
show_in_order(head)
print("\n")
delete_node(2,head)

```
